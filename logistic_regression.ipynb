```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris.data
print(iris["DESCR"])
# print(iris["target"])
```

```python
X = iris["data"]
Y = (iris["target"]==2).astype(np.int32)
```

```python
log_reg = LogisticRegression(C=15)
log_reg.fit(X,Y)
```

```python
X_new = np.linspace(0,3,150).reshape(-1,1)
Y_prob = log_reg.predict_proba(X)
Y_pred = log_reg.predict(X)
```

```python
plt.plot(X, Y_prob[:,1], 'g-',  label='iris-verginica')
plt.plot(X, Y_prob[:,0], 'b--',  label='not-iris-verginica')
plt.xlabel("petal width")
plt.ylabel("probablity of iris virginica")
plt.legend()
```

```python
print(f"train acuracy : {log_reg.score(X,Y)*100}%")
print(f"test acuracy : {log_reg.score(X,Y_pred)*100}%")
print(f"interations : {log_reg.n_iter_}")
```

```python
plt.plot(X_new,Y_prob[:,1], "g-", label="logistic Regression")
plt.scatter(X, Y,  c=Y, cmap="viridis", edgecolors="k")
plt.axhline(0.5, color="red", linestyle="--")
plt.axvline(-(log_reg.intercept_/log_reg.coef_), color="blue", linestyle="-")
plt.xlabel("petal width")
plt.ylabel("probablity of iris virginica")
plt.legend()
plt.show()
```
